On donne la fonction `affiche` qui permet d'afficher une table sous forme de liste de dictionnaire


````python
def affiche(tableau:list,a=0,b=0,numLigne=False) -> None:
    """Affichage d'un tableau"""
    if tableau==[]:
        print("Tableau Vide")
        return
    tab="\t\t"
    if a==0 and b==0:
        a=0
        b=len(tableau) 
    elif b==0:
        b=a
        a=0
    else:
        a=a-1
    if numLigne:
            print("ligne"+"\t",end="")
    for cle in tableau[0]:
        if len(str(cle))>=8:
            print(str(cle)+"\t",end="")
        else:
            print(str(cle)+tab,end="")
    for i in range(a,b):
        print() 
        if numLigne:
                print(str(i+1)+"\t",end="")
        for valeur in tableau[i].values():

            if len(str(valeur))>=8:
                print(str(valeur)+"\t",end="")
            else:
                print(str(valeur)+tab,end="")
    print()
````


!!! note Utilisation de `affiche`
+ `affiche(table)` affiche toutes les lignes.
+ `affiche(table,i,j)` affiche les lignes i inclus  à j inclus.
+ `affiche(table,i,j,True)` affiche les lignes i à j et les numéros de ligne.
