# Exercice 3

```javascript

````python

var AB, AC, BC;

AB = prompt("Donner la valeur AB en cm dans le triangle ABC :");
AC = prompt("Donner la valeur AC en cm dans le triangle ABC :");
BC = prompt("Donner la valeur BC en cm dans le triangle ABC :");

AB = parseFloat(AB);
AC = parseFloat(AC);
BC = parseFloat(BC);
if (AB*AB + AC*AC == BC*BC)
{
console.log("ABC est rectangle en A.");
}
else if (AC*AC + BC*BC == AB*AB)
{
console.log("ABC est rectangle en C.");
}
else if (BC*BC + AB*AB == AC*AC)
{
console.log("ABC est rectangle en B.");
}
else
{
console.log("ABC n'est pas rectangle.");
}
````
